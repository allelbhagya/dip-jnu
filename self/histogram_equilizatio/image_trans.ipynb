```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('image.jpeg')
```

```python
print('Image dimensions:', img.shape)
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

gray = rgb2gray(img)
gray
```

```python
plt.imshow(gray, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
for i in range(len(gray)):
    gray[i] = int(gray[i][0])
gray
```

```python
def step_1(matrix):
    return(max([max(row) for row in gray]))
max_n = step_1(gray)
max_n = int(max_n)
n = max_n.bit_length()
```

```python
gray_freq = [0] * n
for i in range(n):
    gray_freq[i] = np.count_nonzero(gray == i)
print(gray_freq)
```
