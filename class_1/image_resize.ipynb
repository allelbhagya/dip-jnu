## plotting side by side

```python
from PIL import Image

im = Image.open('./img.jpeg')
newsize = (10,10)
im1 = im.resize(newsize)
im1_upscaled = im1.resize(im.size)

total_width = im.width + im1_upscaled.width
new_image = Image.new('RGB', (total_width, im.height))

new_image.paste(im, (0, 0))
new_image.paste(im1_upscaled, (im.width, 0))

new_image
```
