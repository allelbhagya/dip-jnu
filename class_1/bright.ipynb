```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
```

```python
img = mpimg.imread('img.jpeg')
```

```python
img
```

```python
print('image size:', img.shape)
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

gray = rgb2gray(img)
gray
```

```python
plt.imshow(gray, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
gray_new = gray
for i in range(len(gray)):
    gray_new[i] = gray[i]+100
gray_new
plt.imshow(gray_new, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open('img.jpeg').convert('L')

image_array = np.array(image)

brightened_image_array = np.clip(image_array + 50, 0, 255)

brightened_image = Image.fromarray(brightened_image_array.astype(np.uint8))

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image_array, cmap='gray')
ax[0].set_title("Original Image")
ax[0].axis('off')

ax[1].imshow(brightened_image_array, cmap='gray')
ax[1].set_title("Brightened Image (+50)")
ax[1].axis('off')

plt.show()
```
