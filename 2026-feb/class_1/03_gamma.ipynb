```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

def convert_image_to_grayscale(image_path):
    img = Image.open(image_path).convert('L')
    gray_matrix = np.array(img)
    return gray_matrix

def display_image(matrix, cmap='gray'):
    plt.imshow(matrix, cmap=cmap)
    plt.show()

def log_transform(matrix):
    c = 0.5
    gamma = 0.8
    transformed = c * (matrix ** gamma)

    transformed = (transformed / np.max(transformed)) * 255
    return transformed

def process_image(image_path):
    gray_matrix = convert_image_to_grayscale(image_path)
    print("Original Grayscale Matrix:")
    print(gray_matrix)
    
    trans = log_transform(gray_matrix)
    print("Log-Transformed Matrix:")
    print(trans)
    
    display_image(gray_matrix)
    display_image(trans)
```

```python
process_image('cat.jpg')
```
