```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def convert_image_to_grayscale(image_path):
    img = Image.open(image_path).convert('L')
    gray_matrix = np.array(img)
    return gray_matrix

def display_image(matrix, cmap='gray'):
    plt.imshow(matrix, cmap=cmap, vmin=0, vmax=255)
    plt.axis('off')
    plt.show()

def median_filter(matrix, kernel_size=4):
    m, n = matrix.shape
    filtered_matrix = np.zeros((m, n))
    pad_size = kernel_size // 4
    padded_matrix = np.pad(matrix, pad_size)
    for i in range(m):
        for j in range(n):
            region = padded_matrix[i:i+kernel_size, j:j+kernel_size]
            filtered_matrix[i, j] = np.median(region)
    return filtered_matrix

def process_image(image_path):
    gray_matrix = convert_image_to_grayscale(image_path)
    filtered_matrix = median_filter(gray_matrix, kernel_size=3)
    display_image(gray_matrix)
    display_image(filtered_matrix)

process_image('image_noise.png')
```
